## Cleaning

```python
backup=df.copy()
```

```python
df=df[df['year'].str.isnumeric()]                   #checking whether the year column contains any other values than numeric
```

```python
df.head()
```

```python
##changing year dtype from object to integer
df['year']=df['year'].astype(int)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Price']
```

```python
##keeping only numeric values from Price column...excluding "Ask for price"..
df=df[df['Price']!="Ask For Price"]
```

```python
##remove commas " , " by replacing it with an empty string ''.....

df['Price']=df['Price'].str.replace(',','').astype(int)
```

```python
df.info()
```

```python
df['kms_driven']
```

```python
##removing "kms" from kms_driven column by split operation of string .....and then remove commas " , " by replacing it with an empty string ''.....

df['kms_driven']=df['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
##removing non numerical values from kms_driven column...

df=df[df['kms_driven'].str.isnumeric()]
```

```python
df.info()
```

```python
df['kms_driven']=df['kms_driven'].astype(int)
```

```python
df.info()
```

```python
##removing nan values from fuel_type by checking where the nan value is loacted and then excluding that row saving it into the datafrme....

df=df[~df['fuel_type'].isna()]
```

```python
df.info()
```

```python
df['name']
```

```python
backup['name']
```

removing the spam and only keeping the 1st words of name by using split , slice and join operation of strings

```python
df['name']=backup['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df['name']
```

```python
df.head()
```

```python
##now reset the index

df=df.reset_index(drop=True)
```

```python
df
```

```python
df.describe()
```

```python
#since 75% of cars are under 5lac...so checking the outliers...cars which are above 60lacs....and removing that outlier...and the we will reset index


df=df[df['Price']<6e6].reset_index(drop=True)
```

```python
X=df.drop(columns='Price')              #feature columns
y=df['Price']                              #label column
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train ,y_test = train_test_split(X,y,test_size=0.2)
```

## USING OneHotEncoder ( represent categorical variables as numerical values )

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer( (OneHotEncoder(categories=ohe.categories_) , ['name','company','fuel_type']) ,remainder='passthrough' )
print(column_trans)
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

## calculating r2 score (also known as the coefficient of determination is a statistical measure used to evaluate the performance of a regression model. Goodness of Fit)

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(10):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2 , random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

## since the data is very small and the diff betweeen r2 scores are big so we are going with random values / random states of i and creating a list named as score which consist of max r2 score

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2 , random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
